```python
ivestis = input("Įveskite 'taip' arba 'ne': ")
if ivestis.lower() == "taip":
    print("Pasirinkote TAIP")
elif ivestis.lower() == "ne":
    print("Pasirinkote NE")
else:
    print("Nesupratau pasirinkimo")
```

```python
skaicius = int(input("Įveskite skaičių: "))
if skaicius == 0:
    print("Skaičius yra nulis")
elif skaicius % 2 == 0:
    print("Skaičius yra lyginis")
else:
    print("Skaičius yra nelyginis")
```
