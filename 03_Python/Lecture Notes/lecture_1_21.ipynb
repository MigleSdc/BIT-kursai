Nesting lists
A great feature of Python is that they support nesting. This means we can have dardataa structures within data structures. For example: a list inside a list. Let's see how it works:

```python
# Let's make three lists
lst_1 = [1, 2, 3]
lst_2 = [4, 5, 6]
lst_3 = [7, 8, 9]

# Make a list of lists to form a matrix
matrix = [lst_1, lst_2, lst_3]
```

```python
# show
matrix
```

We can again use indexing to grab elements, but now there are two levels for the index. The items in the matrix object and then the items inside that list.

```python
# grab first item in matrix object
matrix[0][0]
```

 Iš žodynų kuriami data frames. Python lentelės, kurios veikia panašiai kaip excel lentelės, jas galima merge'inti ir Join'inti.

```python
transactions = [
    {"region": "LT", "amount": 120.0},
    {"region": "LT", "amount": 80.0},
    {"region": "EE", "amount": 55.0},
    {"region": "LV", "amount": 200.0},
    {"region": "LT", "amount": 40.0},
]

sales_by_region = {}

for t in transactions:
    region = t["region"]
    amount = t["amount"]
    sales_by_region[region] = sales_by_region.get(region, 0) + amount

sales_by_region
```

```python
# zodynas zodyne

products = {
    "SKU-001": {"name": "Kėdė", "price": 49.99, "stock": 120, "category": "Baldai"},
    "SKU-002": {"name": "Stalas", "price": 149.00, "stock": 35, "category": "Baldai"},
    "SKU-005": {"name": "Lempa", "price": 19.90, "stock": 7, "category": "Apšvietimas"},
}
products
```

```python
# reiksmiu paemimas is nested zodyno. reikia naudoti du raktus pirmas: prekes kodas, antras: lauko pavadinimas

sku = "SKU-002"
product_name = products[sku]["name"]
product_stock = products[sku]["stock"]
product_category = products[sku]["category"]
product_name, product_stock, product_category
```

```python
# Ataskaita: isvesti prekes su mazu likuciu
low_stock_thershold = 10
for sku, info in products.items():
    if info["stock"] < low_stock_thershold:
        print("Mazas likutis", sku, info["name"], "likutis", info["stock"])
```

```python
slaptazodis = input("prasome sukurti slaptazodi:")

inputas = input("suveskite savo slaptazodi, kad prisijungtumet prie paskyros")

while inputas != slaptazodis:
    print("nope, netinka, ivesk teisinga")
    inputas = input("suveskite teisinga slaptazodi:")
print("sekmingai prisijungete")
```

```python
slaptazodis = input("prasome sukurti slaptazodi:")

inputas = input("suveskite savo slaptazodi, kad prisijungtumet prie paskyros:")

bandymai = 0
max_bandymai = 3

while inputas != slaptazodis:
    bandymai += 1
    if bandymai > max_bandymai:
        print("pasiekei max skaiciu. patikrinkit ar ne robotas:")
        break
    print("nope, netinka, ivesk teisinga")
    inputas = input("suveskite teisinga slaptazodi:")
if inputas == slaptazodis:
    print("sekmingai prisijungete")
```
