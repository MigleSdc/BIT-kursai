```python
def safe_float(x):
    """Grąžina float arba None, jei konvertuoti nepavyksta."""
    try:
        return float(x)
    except (TypeError, ValueError):
        return None
 
safe_float("12.5"), safe_float("abc"), safe_float(None)
 
```
