```python
vardas = "Jonas"
amzius = 25
print("Vardas:", vardas)
print("Amžius:", amzius)
 
```

```python
#2. formatavimas naudojant f-string
vardas = "Jonas"
amzius = 25
print(f"{vardas} yra {amzius} metu amziaus")
```

```python
#6 keliu formatavimo budu palyginimas
vardas = "Jonas"
amzius = 25
 
# f-string
print(f"{vardas} / {amzius} m.")
 
# format()
print("{} / {} m.".format(vardas, amzius))
 
# klasikinis su kableliais
print(vardas, "/", amzius, "m.")
```

```python
#Tuple fiksuotu reiksmiu rinkinys
# Pvz.: kliento duomenų struktūra
klientas = ("Jonas", 28, "Vilnius")
vardas, amzius, miestas = klientas
print("Vardas:", vardas)
print("Amžius:", amzius)
print("Miestas:", miestas)
```

```python
#Tuple sarase - fiksuota struktura keliose eilutese
#pvz klientu sarasas


#set operacijos - grupiu palyginimas
# Pvz.: aktyvūs ir registruoti klientai
registruoti = {"Jonas", "Ona", "Petras", "Greta"}
aktyvūs = {"Jonas", "Greta"}
 
# Kas registruoti, bet neaktyvūs?
neaktyvūs = registruoti - aktyvūs
print("Neaktyvūs klientai:", neaktyvūs)
```

```python
#Set panaudojimas duomenu analize
# Pvz.: rasti bendrus produktus dviejose rinkose
lt_produktai = {"obuoliai", "bananai", "kriaušės"}
lv_produktai = {"bananai", "kriaušės", "vynuogės"}
 
bendri = lt_produktai & lv_produktai
print("Bendri produktai:", bendri)
```

```python
#Tuple kaip rakto panaudojimas zodyne. dictionary key turi buti unikalus.
# Pvz.: pardavimai pagal (regionas, produktas)
pardavimai = {
    ("Vilnius", "obuoliai"): 150,
    ("Kaunas", "obuoliai"): 200,
    ("Vilnius", "kriaušės"): 180
}
print("Pardavimai Vilniuje (obuoliai):", pardavimai[("Vilnius", "obuoliai")])
```

```python
#dabartine data ir laikas 
import datetime
 
dabar = datetime.datetime.now()
print("Dabartinė data ir laikas:", dabar)
```

```python
#datos kurimas rankiniu budu
gimimo_diena = datetime.date(1990, 7, 19)
print("Gimimo data:", gimimo_diena)
```

```python
#datos konvertavimas is teksto (strptime)
data_tekstu = "2023-12-15"
data_obj = datetime.datetime.strptime(data_tekstu, "%Y-%m-%d").date()
print("Konvertuota data:", data_obj)
```

```python
#biudzeto paskirtsymas pagal menesius be pandas
metinis_biudzetas = 120000
menesio_biudzetas = metinis_biudzetas / 12
 
for men in range(1, 13):
    data = datetime.date(2023, men, 1)
    print(f"{data.strftime('%B')} – {menesio_biudzetas:.2f} EUR")
```

```python
#datetime.timedelta() - laiko skirtumas pvz dienu atimtis
diena1 = datetime.date(2023, 6, 11)
diena2 = datetime.date(2023, 6, 15)
 
skirtumas = diena2 - diena1
print("Dienų skirtumas:", skirtumas.days)
```

```python
#strftime()
dabar = datetime.datetime.today()
print("Metai-mėnuo-diena:", dabar.strftime("%Y-%m-%d"))
print("Laikas (val:min):", dabar.strftime("%H:%M"))
```

```python
gimimo_data = datetime.date(1984, 1, 1)
siandien = datetime.date.today()
amzius = siandien.year - gimimo_data.year
if (siandien.month, siandien.day) < (gimimo_data.month, gimimo_data.day):
    amzius -= 1
print("Amžius:", amzius)
```

```python
#isimtys, ju suvaldymas (su try/except/finally)
#kuo naudingas try/except/finally naudijomas: leidzia pakeisti klaidu pranesimus ispejimu vartotojui. ivykus klaidai, programa nesustoja(apsaugo nuo luzimo).
#po neivykdyto kodo, programa vykdoma toliau
#leidzia nuspresti, ka daryti.(...)
#Kaip suvaldyti kelias isimtis. 
#finally blokas. Kodas, vykdomas nepaisant to, kas vyksta try/except blokuose.
```

```python
#klaidu suvaldymas / error handling su try/ except / finally 
#pagrindinis try / except naudojimas
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Klaida: negalima dalinti iš nulio.")
```

```python
#finally blokas - visada ivykdomas
try:
    f = open("neegzistuojantis_failas.txt")
except FileNotFoundError:
    print("Klaida: failas nerastas.")
finally:
    print("Programa baigė darbą.")
```
