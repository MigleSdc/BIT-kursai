```python
# Yra užsakymo duomenys
order = ("ORD-001", "2024-09-01", 120.50)
 
order
```

```python
#reiksmiu paemimas is tupple pagal indeksa
order_id = order[0]
order_date = order[1]
order_amount = order[2]
 
order_id, order_date, order_amount
```

```python
#tuple unpacking
#naudojamas, kai reikia is karto iskaidyti reiksmes ir kintamuosius ir zinai kokia yra struktura
#tuple unpacking su paskutiniais elementais naudojamas, kai reikalinga tik dalis reiksmiu.
# Yra prekių sąrašas
products = ("Kėdė", "Stalas", "Lempa", "Spinta", "Lentyna")
 
# Reikia paimti paskutinius tris elementus
item1, item2, item3 = products[-3:]
 
item1, item2, item3
 
```

```python
#set unikalios reiksmes
# Yra klientų ID su pasikartojimais
customer_ids = ["C001", "C002", "C001", "C003", "C002"]
 
unique_customers = set(customer_ids)
unique_customers
```

```python
#set pavertimas i list arba tuple naudojamas, kai reikia unikaliu reiksmiu, bet butina seka.
unique_customers_list = list(unique_customers)
unique_customers_tuple = tuple(unique_customers)
 
unique_customers_list, unique_customers_tuple
```

```python
#set operacijos: sajunga, sankirta, skirtumas

#sajungos pvz
campaign_A = {"C001", "C002", "C003"}
campaign_B = {"C002", "C004"}
 
# Klientai, gavę bent vieną kampaniją
campaign_A | campaign_B
```

```python
#klientai, gave abi kampanijas
campaign_A & campaign_B
```

```python
#klientai, gave tik A kampanija
campaign_A - campaign_B
```

String formatting
1.Paprastas print() su kintamaisiais

```python
vardas = "Jonas"
amzius = 25
print("Vardas:", vardas)
print("Amžius:", amzius)
 
```

```python
#2. formatavimas naudojant f-string
vardas = "Jonas"
amzius = 25
print(f"{vardas} yra {amzius} metu amziaus")
```

```python
#3. Formatavimas naudojant str.format()
vardas = "Ona"
amzius = "29"
print ("{} yra {} metu amziaus".format(vardas, amzius))
```

```python
#4. sep panaudojimas (tarpu keitimas) separatoriu nurodai pvz bruksniukas arba taskas
#5. end panaudojimas (pabaigos keitimas)
print("Sveiki,", end=" ")
print("kaip sekasi?")
```

```python
#6 keliu formatavimo budu palyginimas
vardas = "Jonas"
amzius = 25
 
# f-string
print(f"{vardas} / {amzius} m.")
 
# format()
print("{} / {} m.".format(vardas, amzius))
 
# klasikinis su kableliais
print(vardas, "/", amzius, "m.")
```

```python
#Tuple fiksuotu reiksmiu rinkinys
# Pvz.: kliento duomenų struktūra
klientas = ("Jonas", 28, "Vilnius")
vardas, amzius, miestas = klientas
print("Vardas:", vardas)
print("Amžius:", amzius)
print("Miestas:", miestas)
```

```python
#Tuple sarase - fiksuota struktura keliose eilutese
#pvz klientu sarasas


#set operacijos - grupiu palyginimas
# Pvz.: aktyvūs ir registruoti klientai
registruoti = {"Jonas", "Ona", "Petras", "Greta"}
aktyvūs = {"Jonas", "Greta"}
 
# Kas registruoti, bet neaktyvūs?
neaktyvūs = registruoti - aktyvūs
print("Neaktyvūs klientai:", neaktyvūs)
```

```python
#Set panaudojimas duomenu analize
# Pvz.: rasti bendrus produktus dviejose rinkose
lt_produktai = {"obuoliai", "bananai", "kriaušės"}
lv_produktai = {"bananai", "kriaušės", "vynuogės"}
 
bendri = lt_produktai & lv_produktai
print("Bendri produktai:", bendri)
```

```python
#Tuple kaip rakto panaudojimas zodyne. dictionary key turi buti unikalus.
# Pvz.: pardavimai pagal (regionas, produktas)
pardavimai = {
    ("Vilnius", "obuoliai"): 150,
    ("Kaunas", "obuoliai"): 200,
    ("Vilnius", "kriaušės"): 180
}
print("Pardavimai Vilniuje (obuoliai):", pardavimai[("Vilnius", "obuoliai")])
```

```python
#sankirta & arba .intersection()): kas bendro, 
a = {"A", "B", "C"}
b = {"B", "C", "D"}
print(a & b)  # {'B', 'C'}
print(a.intersection(b))
 
```
