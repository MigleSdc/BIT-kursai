8.Apskaičiuokite, kiek procentų liko negrąžintos paskolos (1000 € – 350 € grąžinta).
8. Calculate the percentage of the loan that remains unpaid (€1,000 – €350 repaid).

```python
full_loan = 1000
returned = 350
not_returned = full_loan - returned
percent_not_returned = (not_returned / full_loan) * 100
print("Percent of loan to return is", percent_not_returned, "%")
```

```python
price_with_vat = 121
vat = 0.21
price_without_vat = price_with_vat / (1 + vat)
print("Price without VAT is", price_without_vat, "EUR")
```
